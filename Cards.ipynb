```python
import random
suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
faces= ['jack', 'queen', 'king', 'ace']
numbered= [ 2, 3, 4, 5, 6, 7, 8, 9, 10]

```

```python
suits
```

```python
def draw():
    the_suit = random.choice(suits)
    the_card = random.choice(faces + numbered)
    return f"{the_card} of {the_suit}"
```

```python
print(draw())
```

```python
print(draw())
```

```python
for i in range(5):
    print(draw())
```

```python
for _ in range(5):
    print(draw())
```
