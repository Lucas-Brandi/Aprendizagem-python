```python
FN= "Darius-13-100m-Fly.txt"
```

```python
FOLDER= "swimdata/"
```

```python
with open(FOLDER + FN) as file:
    lines= file.readlines()
```

```python
lines[0]
```

```python
# O método strip() remove os caracteres de nova linha (\n) do final da string, e o método split(",") divide a string em uma lista usando a vírgula como separador.
times= lines[0].strip().split(",")
print(times)
```

```python
first= times[0]
minutes,rest= first.split(":")
seconds, hundredths= rest.split(".")

converted_time= (int(minutes) * 60*100) + (int(seconds) * 100) + int(hundredths) 
print(converted_time)
```

```python
# Convertendo a lista "times" de uma vez
for i in times:
    minutes,rest= i.split(":")
    seconds, hundredths= rest.split(".")
    converted_time= (int(minutes) * 60*100) + (int(seconds) * 100) + int(hundredths) 
    print(f"{i} == {converted_time}")
```
